```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Bengaluru_House_Data.csv")
df.head()
```

```python
df['area_type'].value_counts()
```

```python
df2 = df.drop(['area_type','availability','society','balcony'], axis="columns")
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2 = df2.dropna()
df2.isnull().sum()
```

```python
df2['size'].unique()
```

```python
df2['BHK'] = df2['size'].apply(lambda x: int(x.split(" ")[0]))
df2.head()
```

```python
df2 = df2.drop('size',axis="columns")
df2.head()
```

```python
df2['total_sqft'].unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df2[~df2['total_sqft'].apply(is_float)].head(15)
```

```python
def convert_sqft_num(x):
    token = x.split("-")
    if len(token) == 2:
        return  (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None
    
```

```python
df3 = df2.copy()
df3['total_sqft'] = df3['total_sqft'].apply(convert_sqft_num)
df3 = df3[df3.total_sqft.notnull()]
df3.head()
```

# Feature Enginnering

```python
df4 = df3.copy()
df4.location.nunique()
```

```python
df4.location = df4.location.apply(lambda x: x.strip())
location_stats = df4["location"].value_counts(ascending=False)
location_stats
```

```python
locationstats_less_than_10 = location_stats[location_stats<=10]
locationstats_less_than_10
```

```python
df4.location = df4.location.apply(lambda x: 'other' if x in locationstats_less_than_10 else x )
len(df4.location.unique())
```

```python
df4.head()
```

```python
df5=df4.copy()
df5['price_per_sqft'] = (df5['price']*100000)/df5['total_sqft']
df5.head()
```

# Outlier Removal
### Outliers can be removed by 2 methods. First, if you ask your business manager about the size of a single bedroom and lets say he says the size of one bedroom will be minimum 300sqft. Then we must remove all those datapoints where the size of single bedroom is less than 300sqft.

```python
df5[df5.total_sqft/df5.BHK <300]
```

```python
df6= df5[df5.total_sqft/df5.BHK >=300]
df6.shape
```

## Outlier Removal Using Std Deviation

```python
df6.price_per_sqft.describe()
```

#### The value of price per sqft ranges from 267 to 176470. Price per sqft differ from location to location. So for a particular location we will remove outliers using mean and standard deviation

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df7 = remove_pps_outliers(df6)
df7.shape
```

#### Lets check the price of 2 bhk and 3bhk flats in a particular location

```python
bhk2 = df7[(df7.BHK==2) & (df7.location=='Rajaji Nagar')]
bhk3 = df7[(df7.BHK==3 ) & (df7.location=='Rajaji Nagar')]

plt.scatter(bhk2.total_sqft,bhk2.price, color = "blue")
plt.scatter(bhk3.total_sqft,bhk3.price, color = "red", marker = "+")
plt.xlabel("Total Sqft")
plt.ylabel ("Price")
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('BHK'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('BHK'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df8 = remove_bhk_outliers(df7)
df8.shape
```

```python
bhk2 = df8[(df7.BHK==2) & (df7.location=='Rajaji Nagar')]
bhk3 = df8[(df7.BHK==3 ) & (df7.location=='Rajaji Nagar')]

plt.scatter(bhk2.total_sqft,bhk2.price, color = "blue")
plt.scatter(bhk3.total_sqft,bhk3.price, color = "red", marker = "+")
plt.xlabel("Total Sqft")
plt.ylabel ("Price")
```

```python
df8.bath.unique()
```

```python
df8[df8.bath>df8.BHK+2]
```

```python
df9 = df8[df8.bath<df8.BHK+2]
df9.shape
```

```python
df10 = df9.drop('price_per_sqft' , axis='columns')
df10.head()
```

using pandas dummies to encode location feature

```python
dummies = pd.get_dummies(df10.location)
dummies
```

```python
df11 = pd.concat([df10, dummies] , axis = 'columns')
df11.head()
```

```python
df12 = df11.drop(['location','other'] , axis='columns')
df12.head()
```

```python
X = df12.drop(['price'],axis='columns')
X.head(3)
y = df12.price
y.head(3)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
# Using K fold cross Validation
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5,test_size=0.2, random_state=0)
cross_val_score(LinearRegression(),X,y,cv=cv)
```

```python
# Using Grid Search CV to find out the best model
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```

### from above it is clearly seen that Linear Regression is the best model
